## Data Muning Exercise 

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np

df = pd.read_excel('DFP.xlsx', sheet_name='Data', engine='openpyxl')

print("Column headings:")
print(df.columns)
```

```python
df["amp_or_non_amp"] = np.nan
```

```python
df["story"] = np.nan
```

```python
df["position"] = np.nan
```

```python
print("Column headings:")
print(df.columns)
```

## Creating Story, Position and AMP/NonAMP Columns 


```python
position= ['top', 'middle', 'bottom']
df['position']=df['AD_UNIT_NAME'].apply(lambda x: ''.join([i for i in position if i in x]))

story= ['first', 'second', 'third','fourth','fifth','sixth','seventh','eight','ninth']
df['story']=df['AD_UNIT_NAME'].apply(lambda x: ''.join([i for i in story if i in x]))

df['amp_or_non_amp'] = np.where(df['AD_UNIT_NAME'].str.contains('amp'), 'Amp', 'Nonamp')
```

```python
df
```

```python
print("Column headings:")
print(df.columns)
```

## DAY Column to Map the Day's name in Week 

```python
weeks = {1:'Monday', 2:'Tuesday', 3:'Wednesday',4:'Thursday', 5:'Friday', 6:'Saturday', 7: 'Sunday'}
# Mapping the dictionary keys to the data frame.
df['DAY'] = df['DAY'].map(weeks)
df.head()
```

## Merge Actual_eCPM Data from Actual_eCPM.csv 

```python
df1= pd.read_csv("Actual_eCPM.csv")
df1
```

```python
df = pd.merge(df, df1, on='LINE_ITEM_NAME', how = 'inner')
df
```

```python
df['Actual_eCPM'].values
```

```python
df['Actual_eCPM'] = np.where(df['Actual_eCPM'] == '-', df['eCPM'], df['Actual_eCPM'])
```

```python
df['Actual_Revenue'] = df['Impressions'] * df['Actual_eCPM']
```
